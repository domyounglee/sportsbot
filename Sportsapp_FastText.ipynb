```python
# -*- coding: utf-8 -*-
import re
def onlyHangul(s):
    #s='韓子는 싫고, 한글은 nice하다. English 쵝오 -_-ㅋㅑㅋㅑ ./?!'
    hangul = re.compile('[^ ㄱ-ㅣ가-힣]+') # 한글과 띄어쓰기를 제외한 모든 글자
    # hangul = re.compile('[^ \u3131-\u3163\uac00-\ud7a3]+')  # 위와 동일
    result = hangul.sub('', s) # 한글과 띄어쓰기를 제외한 모든 부분을 제거
    #print (result)
    return result
```

# generating the url vector {url:averaged vector}

```python
from sklearn.preprocessing import normalize
```

```python
kor_video=[]
for vid in video:
    if onlyHangul(vid['description']).split()!=[]:
        kor_video.append(vid)
```

```python
kor_video
```

```python
url_vec={}
for vid in range(len(kor_video)):
    avg_vec=np.zeros( (50) )
    title_li=kor_video[vid]['title'].strip().lower().split()
    print(title_li)
    for word in title_li:
        try:
            avg_vec+=loaded_model.wv.word_vec(word,use_norm=False)
        except:
            continue
    avg_vec/len(title_li)
    avg_vec=normalize(avg_vec.reshape(1,-1))
    avg_vec=np.squeeze(avg_vec)
    url_vec[kor_video[vid]['url']]=avg_vec
```

```python
import pickle 
pickle.dump(url_vec, open( "url_vec.p", "wb" ) )
```

## compute max url 

```python
query=loaded_model.wv.word_vec('초보자를',use_norm=True)+loaded_model.wv.word_vec('푸쉬업',use_norm=True)+loaded_model.wv.word_vec('알려줘',use_norm=True)
query=normalize(query.reshape(1,-1))
query=np.squeeze(query)
max_url=""
max_sim=0
for k,v in url_vec.items():
    dot=v.dot(query)
    if(max_sim<dot):
        max_url=k
        max_sim=dot 
```
